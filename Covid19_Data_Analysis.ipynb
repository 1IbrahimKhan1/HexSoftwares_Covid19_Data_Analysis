```python
#Importing Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the COVID-19 dataset into a DataFrame

df = pd.read_csv(r"C:\Users\khani\Downloads\country_wise_latest.csv", encoding='latin1')
df.head()
```

```python
# Check the number of rows and columns

print("Dataset Shape:")
print(df.shape)

# Display all column names

print("\nColumns:")
print(df.columns)

# Display data types of each column

print("\nData Types:")
print(df.dtypes)
```

```python
# Check missing values
print(df.isnull().sum())

# Remove duplicates
df = df.drop_duplicates()

# Check data types
print(df.dtypes)

# Fill missing values if any
df = df.fillna(0)

# Final shape
print(df.shape)
```

```python
# Generate statistical summary for numerical columns

numeric_df = df.select_dtypes(include=np.number)

numeric_df = numeric_df.replace(
    [np.inf, -np.inf],
    np.nan
)

numeric_df = numeric_df.fillna(0)

numeric_df.describe()
```

```python
# Find top 10 countries with highest confirmed cases

top_cases = df.sort_values(
    by="Confirmed",
    ascending=False
).head(10)

plt.bar(top_cases["Country/Region"],
        top_cases["Confirmed"])

plt.title("Top 10 Countries by Confirmed Cases")
plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# Find top 10 countries with highest deaths

top_deaths = df.sort_values(
    by="Deaths",
    ascending=False
).head(10)

plt.bar(top_deaths["Country/Region"],
        top_deaths["Deaths"])

plt.title("Top 10 Countries by Deaths")
plt.xlabel("Country")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Find top 10 countries with highest recoveries

top_recovered = df.sort_values(
    by="Recovered",
    ascending=False
).head(10)

plt.bar(top_recovered["Country/Region"],
        top_recovered["Recovered"])

plt.title("Top 10 Countries by Recovered Cases")
plt.xlabel("Country")
plt.ylabel("Recovered Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# Group countries by WHO Region
# Calculate total confirmed cases for each region

region_cases = df.groupby(
    "WHO Region"
)["Confirmed"].sum()

region_cases.plot(kind="bar")

plt.title("Confirmed Cases by WHO Region")
plt.xlabel("WHO Region")
plt.ylabel("Confirmed Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# Find countries with highest active cases

top_active = df.sort_values(
    by="Active",
    ascending=False
).head(10)

plt.bar(top_active["Country/Region"],
        top_active["Active"])

plt.title("Top 10 Countries by Active Cases")
plt.xlabel("Country")
plt.ylabel("Active Cases")
plt.xticks(rotation=45)
plt.show()
```

```python
# Scatter plot to analyze relationship
# between new cases and new deaths

plt.scatter(
    df["New cases"],
    df["New deaths"]
)

plt.title("New Cases vs New Deaths")
plt.xlabel("New Cases")
plt.ylabel("New Deaths")
plt.show()
```

```python
# CORRELATION ANALYSIS

# Select only numerical columns

numeric_df = df.select_dtypes(include=np.number)

# Calculate correlation matrix

corr = numeric_df.corr()

# Create heatmap

plt.figure(figsize=(10,8))

plt.imshow(corr)

plt.colorbar()

plt.xticks(
    range(len(corr.columns)),
    corr.columns,
    rotation=90
)

plt.yticks(
    range(len(corr.columns)),
    corr.columns
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Display top 5 countries with highest weekly increase

weekly = df.sort_values(
    by="1 week change",
    ascending=False
).head(5)

print(weekly[
    ["Country/Region",
     "1 week change"]
])
```
